# Merging two datasets(MSR and MVD) with integrated CWE IDs labels

```python
import numpy as np
import pandas as pd
```

```python
def concat_to_df(df):
    df_temp = []
    for chunk in df:
        df_temp.append(chunk)

    df = pd.concat(df_temp,ignore_index = True)
    return df
```

```python
msr_df = pd.read_csv('preprocessed datasets/MSR_new.csv', chunksize=10000, index_col=0)
```

```python
mvd_df = pd.read_csv('preprocessed datasets/MVD.csv', chunksize=10000, index_col=0)
```

```python
msr_df = concat_to_df(msr_df)
mvd_df = concat_to_df(mvd_df)
```

```python
# load dict for later retrieval
import pickle

with open("preprocessed datasets/total_cwe_dict.txt", "rb") as myFile:
    total_cwe_dict = pickle.load(myFile)

print(total_cwe_dict)
```

```python
def check_match_vul_and_CWE_ID(df,cwe_id_col_name):
    count = 0
    for index, row in df.iterrows():
        if row['vul'] == 0 and row[cwe_id_col_name] != 'non-vulnerable':
            print("vul and CWE ID are not matched!!!!!",row['vul'], row['CWE ID'])
            break

    if count==0:
        print("All vul and CWE ID are matched correctly")
        
def check_match_vul_and_label(df):
    count = 0
   
    for index, row in df.iterrows():
        if row['vul'] == 0 and row['label'] != 0:
            print("vul and label are not matched!!!!!",row['vul'], row['label'])
            break
    if count==0:
        print("All vul and label are matched correctly")


            
```

```python
check_match_vul_and_CWE_ID(msr_df,'cwe_id_new')
```

```python
# add new column based on dictionary values
msr_df['label'] = msr_df['cwe_id_new'].apply(lambda x: total_cwe_dict.get(x))
msr_df
```

```python
mvd_pre_df = mvd_df[['code', 'vul', 'CWE ID','label']]

mvd_pre_df.head(10)
```

```python
msr_pre_df = msr_df[['func_before', 'vul', 'cwe_id_new','label']]
msr_pre_df = msr_pre_df.rename(columns={'func_before':'code','cwe_id_new':'CWE ID'})
msr_pre_df
```
